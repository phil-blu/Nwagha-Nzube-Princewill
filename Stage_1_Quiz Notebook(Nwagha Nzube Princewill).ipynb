###### Q9: For the fuel type coal, what is the percentage change in the fuel cost per unit burned in 1998 compared to 1994?

```python
# grouping the dataset based on the fuel type and inputing the fuel desired

fuel_group= fuel_data.groupby('fuel_type_code_pudl')
fuel= input("input the fuel type: ")
fuel_type= fuel_group.get_group(fuel)
fuel_type

# grouping the fuel_type dataframe based on report_year and inputing the first year
year_group= fuel_type.groupby('report_year')
year_1= int(input("input the first year: "))
fuel_year_1= year_group.get_group(year_1)
fuel_year_1

# summing the fuel_qty_burned by the inputed fuel in the first year from fuel_year_1 dataframe
fuel_burned_1= fuel_year_1['fuel_cost_per_unit_burned'].sum()
fuel_burned_1

# grouping the fuel_type dataframe based on report_year and inputing the second year
year_group= fuel_type.groupby('report_year')
year_2= int(input("input the second year: "))
fuel_year_2= year_group.get_group(year_2)
fuel_year_2

# summing the fuel_qty_burned by the inputed fuel in the second year from fuel_year_2 dataframe
fuel_burned_2= fuel_year_2['fuel_cost_per_unit_burned'].sum()
fuel_burned_2

# calculating the %change in the qty of fuel burned between the two years
percentage_mean_difference= ((fuel_burned_2 - fuel_burned_1)/fuel_burned_1) * 100
percentage_mean_difference
```

#### Q.10: Which year has the highest average fuel cost per unit delivered?

```python
# importing pandas as pd

import pandas as pd

# grouping the fuel_dataset based on the report_year and converting to a dataframe
year_group= fuel_data['report_year']
year_group= pd.DataFrame(year_group)
year_group

# grouping the fuel_dataset based on the fuel_cost_per_unit_delivered and converting to a dataframe
fuel_delivered= fuel_data['fuel_cost_per_unit_delivered']
fuel_delivered= pd.DataFrame(fuel_delivered)
fuel_delivered

# concatenating both dataframes
group= pd.concat([year_group, fuel_delivered], axis= 1)
group
```

```python
year_group= group.groupby('report_year')
year= int(input('input the year: '))

each_year= year_group.get_group(year)
avg_cost= each_year['fuel_cost_per_unit_delivered'].mean()
avg_cost
```
